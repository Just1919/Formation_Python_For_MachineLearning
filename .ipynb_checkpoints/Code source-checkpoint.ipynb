## Inerpolate

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,10)
y=x**2
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind='linear')
new_x=np.linspace(0,10,30)
result=f(new_x)
plt.scatter(x,y)
plt.scatter(new_x,result,c="red")
```

## Optimize

```python
x=np.linspace(0,2,100)
y=1/3*x**3-3/5*x**2+2+np.random.randn(x.shape[0])/20
plt.scatter(x,y)
```

```python
def f(x,a,b,c,d):
    return a*x**3+b*x**2+c*x+d
```

```python
from scipy import optimize
```

```python
params,parm_cuv=optimize.curve_fit(f,x,y)
plt.scatter(x,y)
plt.plot(x,f(x,params[0],params[1],params[2],params[3],),c='r')
```

## Minimisation

```python
def f(x):
    return x**2 + 15*np.sin(x)
    
```

```python
x=np.linspace(-10,10,100)
plt.plot(x,f(x))
```

```python
x0=-5
result=optimize.minimize(f,x0=x0).x
plt.scatter(result,f(result),s=100,c='r',zorder=1)
plt.scatter(x0,f(x0),marker='+',s=300,c='g',zorder=1)
plt.plot(x,f(x),lw=3,zorder=-1)
```

```python
def f(x):
 return np.sin(x[0])+np.cos(x[0]+x[1])*np.cos(x[0])
```

```python
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)
x,y=np.meshgrid(x,y)
plt.contour(x,y,f(np.array([x,y])),20)

x0=np.zeros((2,1))
plt.scatter(x0[0],x0[1],s=200,marker="+",c="red")

result=optimize.minimize(f,x0=x0).x
plt.scatter(result[0],result[1],s=200,marker="*",c="black")
```

## Transformation de Fourier

```python
x=np.linspace(0,30,1000)
y=3*np.sin(x)+2*np.sin(5*x) + np.sin(10*x)
plt.plot(x,y)
```

```python
from scipy import fftpack
```

```python
fourier=fftpack.fft(y)
power=np.abs(fourier)
frequences=fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences),power)
```

```python
#Exemple
x=np.linspace(0,30,1000)
y=3*np.sin(x) + 2*np.sin(5*x) + np.sin(10*x)+ np.random.randn(x.shape[0])*10
plt.plot(x,y)
```

```python
from scipy import fftpack
```

```python
fourier=fftpack.fft(y)
power=np.abs(fourier)
frequences=fftpack.fftfreq(y.size)
plt.plot(np.abs(frequences),power)
```

```python
fourier[power<600]=0
```

```python
plt.plot(np.abs(frequences),np.abs(fourier))
```

```python
filtered_signal=fftpack.ifft(fourier)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y,lw=0.5)
plt.plot(x,filtered_signal,lw=3)
plt.show()
```

## Sries temporelles et analyse de Bitcoin

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bitcoin=pd.read_csv("BTC-EUR.csv",index_col="Date",parse_dates=True)
```

```python
bitcoin.head()
```

```python
bitcoin["Close"].plot(figsize=(9,6))
```

```python
bitcoin["2019":"2020"]["Close"].plot()
```

### RESAMPLE

```python
bitcoin.loc["2019","Close"].resample("M").plot()
plt.show()
```

```python
bitcoin.loc["2019","Close"].resample("2W").mean().plot()
plt.show()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc["2019","Close"].plot()
bitcoin.loc["2019","Close"].resample("M").mean().plot(label="moyenne par mois",lw=3,ls=':',alpha=0.8)
bitcoin.loc["2019","Close"].resample("W").mean().plot(label="moyenne par semaine",lw=2,ls='--',alpha=0.8)
plt.legend()
plt.show()
```

## AGGREGATE

```python
bitcoin.loc["2019","Close"].resample("M").agg(["min","max","mean","std"])
```

```python
m=bitcoin.loc["2019","Close"].resample("M").agg(["min","max","mean","std"])
plt.figure(figsize=(14,6))
m["mean"]["2019"].plot(label="moyenne par semaine")
plt.fill_between(m.index,m["max"],m["min"],alpha=0.2,label="min-max par mois")
plt.legend()
plt.show()
```

## MOVIENG AVERAGE

```python
plt.figure(figsize=(12,8))
bitcoin.loc["2019-09","Close"].plot()
bitcoin.loc["2019-09","Close"].rolling(window=7).mean().plot(label="Moving Average",c="red")
bitcoin.loc["2019-09","Close"].rolling(window=7,center=True).mean().plot(label="Moving Average")
bitcoin.loc["2019-09","Close"].ewm(alpha=0.6).mean().plot(label="Moving Average",c="black")
plt.legend()
plt.figure(figsize=(14,6))
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc["2019-09","Close"].plot()
for i in np.arange(0.2,1,0.2):
    bitcoin.loc["2019-09","Close"].ewm(alpha=i).mean().plot(label=f'ewm{i}',ls="--",alpha=0.8)
plt.legend()
plt.show()
```
